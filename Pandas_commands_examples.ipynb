```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
univ = pd.read_csv('Universities.csv')
```

```python
univ
```

```python
univ.shape
```

```python
plt.hist(univ['Accept'])
plt.xlabel('Acceptance')
plt.ylabel('count')
plt.title('Histogram of acceptance')
plt.show()
```

```python
plt.hist(univ['SAT'])
plt.xlabel('SAT')
plt.ylabel('count')
plt.title('Histogram of SAT')
plt.show()
```

```python
sn.histplot(univ['Accept'])
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sn.distplot(univ['Accept'])
plt.show()
```

```python
plt.plot(univ['Accept'],univ['Top10'])
plt.show()
```

```python
plt.plot(univ['Accept'],univ['Top10'], 'rs')
plt.xlabel('Accept')
plt.ylabel('Top10')
plt.title('Scatterplot of Accept vs Top10')
plt.show()
help(plt.plot)  # it will help us to learn graph
```

```python
sn.pairplot(univ)
```

```python
sn.pairplot(univ.iloc[:,0:3])
```

```python
plt.boxplot(univ['Accept'])
plt.show()
```
